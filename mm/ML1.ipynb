```python
import pandas as pd

df=pd.read_csv('covid19.csv',index_col=0)
df.head(250)
```

```python
import missingno
missingno.matrix(df)
```

```python
df_missing = pd.DataFrame(df.isnull().sum()/df.shape[0],columns=['missing_rate']).reset_index()#
df_missing.sort_values(by='missing_rate',ascending=False)[:15]
#if was writting [:] will get all the columns
```

```python
missing_col = ["NewCases","TotalDeaths","TotalCases","Country,Other"]
for col in missing_col:
    df[col] = df[col].fillna(df[col].mode())
print(df.to_string())
```

```python
df_missing = pd.DataFrame(df.isnull().sum()/df.shape[0],columns=['missing_rate']).reset_index()#
df_missing.sort_values(by='missing_rate',ascending=False)[:15]
```
